```python
import pandas as pd

df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
# View null values
df.isnull().sum()
```

```python
# Replacing the null values
# Mean: It can get affected by outliers and can skew the distribution.
# Median: It is robust to outliers.
```

```python
# It not a good step to replace all the values with median
# Because ratings may vary, Footwear might show different trends than clothing category that might be bias in the dataset.
```

```python
# So we will impute median with the median values of each specific category
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Replace every column names to lower case and underscore
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
# Create a columns for age group
labels = ['young_adult', 'adult', 'middle_aged', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head()
```

```python
# Purchase_frequency_days
# Textual frequency into numbers
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90,
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
# Both columns have same value, so its redundant column
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Lets remove one column
df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine

# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details
username = "postgres"      # default user
password = "root" 
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "customer_behavior"    # the database created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "customer"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
